```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
from sklearn.preprocessing import normalize
```

```python
x = new_df_jobs.drop(['y'],axis=1)
y = new_df_jobs['y']
```

```python
array = normalize(x.values, axis=0, norm='max')
print(array)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42)
```

```python
# creting the decision tree model
model = DecisionTreeClassifier()
```

```python
dTree = model.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
x_predict = model.predict(x_test)
```

```python
accuracy_score(x_predict,y_test)
```

```python
model=CausalModel(
        data = new_df_jobs,
        treatment= "t",
        outcome= "y",
        common_causes=['x1','x2','x3','x4','x5','x6','x7','x8','x9','x10','x11','x12','x13','x14','x15','x16','x17','e'],
        )
```

```python
# View model
model.view_model()
from IPython.display import Image, display
display(Image(filename="causal_model.png"))
```

```python
#Generate estimand i.e. recipe for calculating causal effect
identified_estimand= model.identify_effect(proceed_when_unidentifiable=True)
print(identified_estimand)
```

```python
#Compute causal effect using metaleaner

identified_estimand_experiment = model.identify_effect(proceed_when_unidentifiable=True)

from sklearn.ensemble import RandomForestRegressor
metalearner_estimate = model.estimate_effect(identified_estimand_experiment,
                                method_name="backdoor.econml.metalearners.TLearner",
                                confidence_intervals=False,
                                method_params={"init_params":{
                                                    'models': RandomForestRegressor()
                                                    },
                                               "fit_params":{}
                                              })
print(metalearner_estimate)
```

```python
# print histogram of causal effects for each sample
plt.hist(metalearner_estimate.cate_estimates)
```
